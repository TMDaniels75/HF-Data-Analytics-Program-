```python
import numpy as np
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

```python
df
```

```python
df[['School DBN','School Level','School Name', 'Regents Exam', 'Year', 'Mean Score']]
```

```python
df2=df[['School DBN','School Level', 'School Name','Regents Exam', 'Year', 'Mean Score']]
```

```python
df2
```

```python
borough_map = {
    'M': 'Manhattan',
    'Q': 'Queens',
    'K': 'Brooklyn',
    'R': 'Staten Island',
    'X': 'Bronx'
}
```

```python
df2["Borough"] =df2["School DBN"].str[2].map(borough_map)
```

```python
df2
```

```python
drop_cols=['School DBN']
```

```python
df2.drop(drop_cols,axis=1,inplace=True)
```

```python
df2
```

```python
df2 = df2[~df2["Mean Score"].str.contains("s", case=False, na=False)]
```

```python
df2
```

```python
df2['Year'].unique()
```

```python
df_2015 = df2[df['Year'] == 2015]
df_2016 = df2[df['Year'] == 2016]
df_2017 = df2[df['Year'] == 2017]
```

```python
df_2015
```

```python
df_2015.sort_values('Mean Score', ascending=False).head(10)
```

```python
top10 = df_2015.sort_values("Mean Score", ascending=False).head(10)

import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

labels = top10["Mean Score"] + " (" + top10["Borough"] + ")"

plt.bar(labels, top10["Mean Score"])

plt.xticks(rotation=45, ha="right")
plt.xlabel("Borough")
plt.ylabel("Mean Score")
plt.title("Top 10 2015 Exam Scores by Borough")

plt.show()
```
